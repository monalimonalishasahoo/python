```python
def average(numbers):
    total = 0
    for num in numbers:
        total = total + num
        avg = total/len(numbers)
        return avg
    
def main():
    prices = [29, 21, 55, 10]
    print("Prices :",prices)
    result = average(prices)
    print(result)
    
main()
```

```python
def average(numbers):
    total = 0
    for num in numbers:
        total = total + num
        avg = total/len(numbers)
        return avg

def main(order_goal):
    prices = [29, 21, 55, 10]
    result = average(prices)
    print(order_goal)
    print(result)
    
order_goal = 25
main(order_goal)
```
