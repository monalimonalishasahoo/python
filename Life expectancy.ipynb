```python
#prints a dataframe which shows no. of nans from each column.
a=list(ds.columns)
b=[]
for i in a:
    c=ds[i].isnull().sum()
    b.append(c)
null=pd.DataFrame({'Feature name':a,'no. of Nan':b})
null
```

```python
#filling the nans with the mean value of the columns
for column in ds.columns:
    if ds[column].isnull().sum() != 0:
        ds[column].fillna(ds[column].mean(), inplace=True)
```

```python
#showing the mean life expectancy of each country
fig = plt.subplots(figsize=(25, 10))
g1 = ds.groupby("Country")["Life expectancy "].mean()
g1.plot(kind="bar", alpha=0.3, ec="k", color="yellow")
plt.ylabel("Mean life expectancy")
plt.show()
```

```python
#Country and Status columns aren't numerical values ,so convert them into numericals.
print(ds["Country"].unique())
letters = ds["Country"].unique()
numbers = list(range(len(ds["Country"].unique())))
d = dict(zip(letters, numbers))
ds["Country"] = ds["Country"].map(d)
print(ds["Country"].unique())

print(ds["Status"].unique())
mappingstatus = {"Developing": 0, "Developed": 1}
ds["Status"] = ds["Status"].map(mappingstatus)
print(ds["Status"].unique())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV
```

```python
#split the data
x = ds.drop("Life expectancy ", axis=1)
y = ds["Life expectancy "]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.20,random_state=500)
```

```python
Linear_model= LinearRegression()
Linear_model.fit(xtrain,ytrain)
```

```python
predictions=Linear_model.predict(xtest)
```

```python
r2_score(ytest,predictions)
```
